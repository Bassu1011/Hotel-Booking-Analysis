# **Data cleaning**


*   Data cleaning is the most important step in the EDA process. And this is the most time consuming process.
*   Data cleaning is the process of fixing or removing incorrect, corrupted, incorrectly formatted, duplicate, or incomplete data within a dataset.

*  Data cleaning includes fallowing steps.
  1. Remove duplicates or irrelevent observations
  2. Handle missing data 
  3. Fix structural errors
  4. Adding or removing of some columns for analysis
  5. Validate and QA




**Step - 1: Removing duplicates.**

```python
duplicate_values=df.duplicated()
duplicate_values.value_counts()
```

```python
#original shape of the data
df.shape
```

True means Duplicate data. there are total 31994 duplicate rows among 119390 rows in the dataset

```python
#now drop the duplicate rows from the dataset and check the shape
df.drop_duplicates(inplace=True)
df.shape
```

Now dataset reduced to 87396 row after dropping the duplicate rows

**Step 2: Handling missing data**

```python
#check null values
df.isnull().sum().sort_values(ascending=False)[:10]
```

**Inference :**

As we see above there four columns having null values, in them children having the least null value while company and agent having more nul values.

As company and agent columns have company number and agent number as data. So there are some cases where customer booked hotel directly without contacting any agent or hotel booking company so we will replace that value with 0, as well as in children column their might be a chance that data has been not been entered so we will replace that too with 0.

```python
#replace null values of company, agent and children with 0
df[['children','company','agent']] = df[['children','company','agent']].fillna(0)
```

Now in country column we have missing values some people might have not filled data of country from which country they came, So we will replace those null values with 'Not Specified'.

```python
#replacing null values in country columns with 'Not Specified'.
df['country'].fillna('Not Specified',inplace=True)
```

```python
#Now let's check all the missing values are replaced or there are still any.
df.isnull().sum().sort_values(ascending=False)[:10]
```

**Step 3: Fix structural errors**

In this process columns are converted to proper datatype or format.

As we see in the information of the dataset, some coulumns don't have proper data type or format, now we convert them to the proper format or datatype.

```python
#convert dtype of children, agent, company from float to int
df[['children','company','agent']] = df[['children','company','agent']].astype('int64')
```

```python
#as reservation_status_date is in object dtype so we will convert it into datetime
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'],format = '%Y-%m-%d')
```

**Step 4: Adding or removing of some columns for analysis**

```python
#add stays_in_week_nights and stays_in_weekend_nights to create total_stay column.
df['total_stay'] = df['stays_in_week_nights']+df['stays_in_weekend_nights']
```

```python
#add adults, children and babies to create total_people column
df['total_people'] = df['adults']+df['children']+df['babies']
```
